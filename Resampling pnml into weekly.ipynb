```python
import pandas as pd
import datetime
tp_02=pd.read_csv("tp02waterdata.csv")
tp_04=pd.read_csv("tp04waterdata.csv")
tp_05=pd.read_csv("tp05waterdata.csv")
print(tp_02)
```

```python
#A week starts on Monday and ends on Sunday (I know that you are probably used to the
#concept that a new week starts on Sunday, but I also checked the dates and since the first date
#we have is Monday and the last one is on Sunday I think for consistency concern we'd probably
#be better off with the week starting on Monday and ending on Sunday so that we don't have to
#create a new week for that one data and we would not be missing one data point for the first week)
print(tp_02.Sample_Date.dtype)
```

```python
#pilot run 
start_date="2018-05-14"
end_date="2018-05-21"
mask=(tp_02['Sample_Date'] > start_date) & (tp_02['Sample_Date'] <= end_date)
print(tp_02.loc[mask])

start_date=datetime.datetime.strptime(start_date, '%Y-%m-%d')
print(start_date)
print(start_date+datetime.timedelta(days=7))
end_date=start_date+datetime.timedelta(days=7)

mask=(tp_02['Sample_Date'] > str(start_date)) & (tp_02['Sample_Date'] <= str(end_date))
print(tp_02.loc[mask])

#This works
```

```python
#adding week column, with 2018-05-14 to 2018-05-20 being the first week (being assigned value 1) and so on
tp_02.drop(columns=['Unnamed: 0'],inplace=True)
print(tp_02)
tp_02_weekly=pd.DataFrame(columns=['Chemical','Concentration_ng_L_Error','Concentration_ng_L_Value','Day','Mass_Load_g_d_Error','Mass_Load_g_d_Value','Month','Population_Normalized_Mass_Lo_1','Population_Normalized_Mass_Load','Sample_Date','Site','Week'])
for i in range(71): #70 weeks in total in pilot dataset
    start_date="2018-05-14"
    start_date=datetime.datetime.strptime(start_date, '%Y-%m-%d')
    start_date=start_date+datetime.timedelta(days=7*i)
    end_date=start_date+datetime.timedelta(days=7)
    mask=(tp_02['Sample_Date'] > str(start_date)) & (tp_02['Sample_Date'] <= str(end_date))
    #print(tp_02.loc[mask])
    temp=tp_02.loc[mask]
    
    if not temp.empty:
        temp.loc[:,'Week']=i
        #print(temp)
        tp_02_weekly=pd.concat([tp_02_weekly, temp], ignore_index=True)
    #if not temp.empty:
        
#print(tp_02)
print(tp_02_weekly)
```
