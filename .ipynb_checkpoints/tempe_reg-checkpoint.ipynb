#### PNML Regression

```python
#with both fe
f = 'pnml ~ MedFacilitiesCount2mi + PharmCount2mi + DrugDropCount2mi + NaloxoneDistribCount2mi +C(year_month) +C(Site)'
res = smf.ols(formula=f, data=mt).fit(drop_first=True)
res.summary()
```

```python
plt.hist(res.resid)
```
