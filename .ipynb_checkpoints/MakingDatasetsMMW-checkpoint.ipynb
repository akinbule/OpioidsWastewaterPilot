```python
url = 'https://services.arcgis.com/lQySeXwbBg53XWDi/arcgis/rest/services/Wastewater_Study_Areas/FeatureServer/1/query?f=json&outFields=*&returnGeometry=false&where=Site%3D%27TP02%27'
resp = requests.get(url)
print(resp)

data_raw = resp.json()
data_raw.keys()
```

```python
data_raw['features']
tp02_data = pd.io.json.json_normalize(data_raw['features'])
tp02_data.columns = ["Chemical","Concentration_ng_L_Error","Concentration_ng_L_Value","Day","Mass_Load_g_d_Error","Mass_Load_g_d_Value","Month","OBJECTID","Population_Normalized_Mass_Lo_1","Population_Normalized_Mass_Load","Sample_Date","Site"]
```

```python
times = []
for t in tp02_data["Sample_Date"]:
    t1 = str(t)
    t2 = t1[0:10]
    t2 = int(t2)
    time = datetime.fromtimestamp(t2)
    times.append(time)
tp02_data["Sample_Date"] = times
tp02_data.head()
```

```python
tp02_data.drop(['OBJECTID'], axis=1, inplace=True)
```

```python
tp02_data.head()
```

```python
tp02_data.to_csv(r'DataFiles/tp02waterdata.csv')
```

**This is to get the EMS Opioid call data aggregated by month-year for each tract.

```python
infile = open('DataFiles/emsPulledData.pkl','rb')
ems_test = pkl.load(infile)
infile.close()
```

```python
from time import strptime

ems_test['Weekday'] = ems_test['Weekday'].apply(lambda x: strptime(x,'%A').tm_wday)
ems_test['Month'] = ems_test['Month'].apply(lambda x: strptime(x,'%B').tm_mon)

ems_test["Date"] = ems_test["Month"].map(str) + '/' + ems_test['Weekday'].map(str) + '/' + ems_test["Year"].map(str)

ems_test.head()
```

```python
times = []
for t in ems_test["Date"]:
    time = datetime.strptime(t, "%m/%w/%Y")
    times.append(time)
ems_test["Date"] = times
ems_test['Month_Year'] = pd.to_datetime(ems_test['Date']).dt.to_period('M')
```

```python
ems_test.drop(['Date'], axis=1, inplace=True)
ems_test.head()
```

```python
ems_test.drop(['Sample_Date'], axis=1, inplace=True)
ems_test.drop(['Date'], axis=1, inplace=True)
ems_test.head()
```

```python
#All EMS Opioid Call data in Tempe aggregated by Month/Year
with open('DataFiles/emsPulledData_0118_0819.pkl', 'wb') as pkl_file:
    pkl.dump(ems_test, pkl_file)
```

```python
#CREATING EMS DATAFRAME AGGREGATED BY MONTH FOR EACH TRACT
ems_df_full_mnyr = pd.DataFrame()
#ems_df_full.columns = odf.columns

odf = tempe_tract_df

for i in range(len(tempe)):
    df = ems_test[ems_test.within(odf.loc[i,'geometry'])]
    ems_monthly = df.filter(['Month_Year', 'Opioid_Use']).groupby('Month_Year').count()
    num = []
    for m in ems_monthly['Opioid_Use']:
        n = m
        num.append(m)
    months = list(np.unique(df['Month_Year']))
    x = len(ems_monthly['Opioid_Use'])
    t = pd.DataFrame(odf.loc[i]).T
    e_m = pd.DataFrame(np.repeat(t.values,x,axis=0))
    e_m.columns = t.columns
    e_m['Month'] = months
    e_m['Opioid_Use_EMS'] = num
    ems_df_full_mnyr = ems_df_full_mnyr.append(e_m)

ems_df_full_mnyr
```

```python
with open('DataFiles/emsCensusTract0118_0819.pkl', 'wb') as pkl_file:
    pkl.dump(ems_df_full_mnyr, pkl_file)
```

**This is to get EMS data aggregated strictly by month (so 2018 and 2019 together) by census tract

```python
#CREATING EMS DATAFRAME AGGREGATED BY MONTH [NOT MONTH/YEAR] FOR EACH TRACT

ems_df_full = pd.DataFrame()
#ems_df_full.columns = odf.columns

odf = tempe_tract_df

for i in range(len(tempe)):
    df = ems_allc[ems_allc.within(odf.loc[i,'geometry'])]
    ems_monthly = df.filter(['Month_Sort', 'Opioid_Use']).groupby('Month_Sort').count()
    num = []
    for m in ems_monthly['Opioid_Use']:
        n = m
        num.append(m)
    months = list(np.unique(df['Month_Sort']))
    x = len(ems_monthly['Opioid_Use'])
    t = pd.DataFrame(odf.loc[i]).T
    e_m = pd.DataFrame(np.repeat(t.values,x,axis=0))
    e_m.columns = t.columns
    e_m['Month'] = months
    e_m['Opioid_Use_EMS'] = num
    ems_df_full = ems_df_full.append(e_m)

ems_df_full
```

```python
with open('DataFiles/emsCensusTract_Month.pkl', 'wb') as pkl_file:
    pkl.dump(ems_df_full, pkl_file)
```
