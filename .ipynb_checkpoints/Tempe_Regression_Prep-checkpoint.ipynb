### Pull Variables from the Census and Merge by Geometry
This will allow us to pull demographic predictors from the census. Everything in the var list below we are essentially considering as part of our X matrix, while our y is either PNML measure in that geometry or the count of EMS calls

All code below is a functional example that can you get started, but there is probably tons of ways to do this a) more efficiently and b) more thoroughly. Please experiment! 

```python
var = ['NAME', 'B00001_001E', 'B01001_002E', 'B02001_002E', 'B02001_003E', 'B02001_004E', 'B02001_005E', 'B02001_006E',
      'B02001_007E', 'B02001_008E', 'B02001_009E', 'B02001_010E', 'B05001_006E', 'B19001_001E', 'B22003_002E',
       'B08121_001E','B22003_003E', 'B19326_001E', 'B19013_001E']

# data dictionary and examples of variables to pull https://api.census.gov/data/2016/acs/acs5/variables.html
#B00001_001E - total pop
#B01001_002E - total hh 
#B02001_002E to "...10E" - race data 
#B05001_006E - total non citizen 
#B19001_001E - HHI (check that this is right?)
#B22003_002E - recieved food stamps 
#B22003_003E - recieved food stamps and below poverty level
#B08121_001E - median earnings (check that this is right?)
#B19326_001E
#B19013_001E - income?
```

```python
#using census api wrapper to get the ACS 5 year estimates for all census tracts in Arizona
xtr = c.acs5.get(var, geo={'for': 'tract:*',
                       'in': 'state:{} county:013'.format(states.AZ.fips)})
```

```python
cens = pd.DataFrame(xtr)
```

```python
#restrict to just Maricopa county
cens[cens['tract']=='319104']
```

```python
tempe_full = tempe.merge(cens, left_on='TRACTCE', right_on='tract')
```

```python
tempe_full.head()
```

```python
#remove outlier census tract
tempe = (tempe_full[tempe_full['tract'] != '113802'])
```

```python
#replot to ensure integrity
fig, ax = plt.subplots(figsize=(10,10))
plt.title('Collection Areas Overlaid on City Boundary and Census Tracts in or partially in Tempe, AZ', size=18)
tempe.boundary.plot(ax=ax, color='grey')
bdry.boundary.plot(ax=ax, color='green')
sa[sa['Site'] == 'TP02'].boundary.plot(ax=ax, color='red')
sa[sa['Site'] == 'TP04'].boundary.plot(ax=ax, color = 'yellow')
sa[sa['Site'] == 'TP05'].boundary.plot(ax=ax, color = 'blue')
ems.plot(ax=ax)
```

```python
#simple chloreograph from column overlaid on our geometry
#heat map of poverty level & food stamps by census tract
fig, ax = plt.subplots(figsize=(10,10))
tempe.plot(column='B22003_003E', ax=ax, legend=True)
sa[sa['Site'] == 'TP02'].boundary.plot(ax=ax, color='red')
sa[sa['Site'] == 'TP04'].boundary.plot(ax=ax, color = 'yellow')
sa[sa['Site'] == 'TP05'].boundary.plot(ax=ax, color = 'blue')
plt.title('Census Tract Heat Map: Below Poverty and Recieved Food Stamps')
```
